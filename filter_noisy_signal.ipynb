### Step 1: Setup and Generate the Noisy Signal

```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
```

```python
#Generate a sinusoidal signal with a frequency of 1 Hz and amplitude of 1
# Sampling parameters
fs = 100  # Sampling frequency in Hz
t = np.linspace(0, 10, fs * 10, endpoint=False)  # Time vector for 10 seconds

# Original signal
freq = 1  # Frequency of the sinusoidal signal
amplitude = 1  # Amplitude of the sinusoidal signal
original_signal = amplitude * np.sin(2 * np.pi * freq * t)
```

```python
#Introduce Gaussian noise with zero mean and a standard deviation of 0.5
noise = np.random.normal(0, 0.5, t.shape)
noisy_signal = original_signal + noise
```

## Step 2: Design a Low-pass Filter

```python
#Choose a filter type and determine the filter order and cutoff frequency
# Filter design parameters
cutoff_freq = 2  # Cutoff frequency in Hz (just above the signal frequency)
filter_order = 4  # Filter order

# Design a Butterworth low-pass filter
b, a = butter(filter_order, cutoff_freq / (0.5 * fs), btype='low')
```

## Step 3: Apply the Filter

```python
#Apply the designed filter to the noisy signal
filtered_signal = lfilter(b, a, noisy_signal)
```

## Step 4: Plot Results

```python
#Plot the original signal, noisy signal, and filtered signal
# Plot the signals
plt.figure(figsize=(12, 8))

# Plot original signal
plt.subplot(3, 1, 1)
plt.plot(t, original_signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot noisy signal
plt.subplot(3, 1, 2)
plt.plot(t, noisy_signal)
plt.title('Noisy Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot filtered signal
plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal)
plt.title('Filtered Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
